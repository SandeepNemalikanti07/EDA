```python
selected_features=['Debt ratio %',
                'Current Liability to Assets',
                "ROA(A) before interest and % after tax",
                "ROA(B) before interest and depreciation after tax",
                "ROA(C) before interest and depreciation before interest",
                'Borrowing dependency',
                'Net Income to Total Assets',
                'Gross Profit to Sales']
```

```python
outlier_summary = []
for col1 in selected_features:
    q1 = bank[col1].quantile(0.25)
    q3 = bank[col1].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = ((bank[col1] < lower) | (bank[col1] > upper)).sum()

    outlier_summary.append({
        "feature": col1,
        "outlier_count": outliers,
        "lower_bound": lower,
        "upper_bound": upper
    })

outlier_df = pd.DataFrame(outlier_summary)
outlier_df["percentage"] = (outlier_df["outlier_count"] / len(bank)) * 100
outlier_df.sort_values("percentage", ascending=False)
```

```python
#In bankruptcy data, extreme values (very high debt, very low profit) are often actual distressed firms, not errors.
#If removed → model loses ability to detect high-risk cases
```

```python

num_df = bank.select_dtypes(include=['int64','float64'])
corr = num_df.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
good_corr = (
    upper.stack()
    .reset_index()
    .rename(columns={0: 'correlation'})
    .query("correlation >= 0.85")
    .sort_values(by='correlation', ascending=False)
)

print(good_corr.head(10))
```

```python

top_features = (
    bank.corr(numeric_only=True)['Bankrupt?']
    .abs()
    .sort_values(ascending=False)
    .head(10)
    .index
)

# correlation matrix of those features
top_corr_matrix = bank[top_features].corr()

plt.figure(figsize=(8,6))
sns.heatmap(top_corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Top Feature Correlation Matrix")
plt.show()
```

1. Bankruptcy Is a Rare Event (Highly Imbalanced)

Most bankruptcy datasets show that bankrupt companies make up a very small fraction of the data (often <5%). This means:

Overall accuracy metrics are misleading; focus on recall for the bankrupt class.

Models and business decisions must prioritize detection of rare bankruptcy signals.

Insight: Risk monitoring systems should be tuned for early detection of distress rather than overall classification performance.

2. Key Financial Ratios Correlate with Bankruptcy Risk

Certain financial ratios are consistently more predictive:

Liquidity Ratios (e.g., Current Ratio, Quick Ratio)

Profitability Ratios (e.g., ROA, Net Profit Margin)

Leverage Ratios (e.g., Debt-to-Equity)
Companies with lower liquidity, poor profitability, and high leverage are more likely to fail.

Insight: Regularly tracking these ratios can serve as an early warning indicator for financial distress.

3. Outliers and Extreme Values Matter

EDA typically reveals outliers in financial ratios (very high debt, very low cash reserves). These outliers:

May signal financial stress before bankruptcy occurs.

Should not be removed before modeling because they carry meaningful risk signals.

Insight: Extreme financial values should be flagged for review, not discarded.

4. Feature Correlation Identifies Redundancy

Many financial metrics are correlated (e.g., multiple measures of profitability or liquidity). Correlation analysis helps:

Reduce dimensionality.

Focus on unique sources of risk.

High correlation between related features suggests overlapping risk signals.

Insight: Simplify dashboards and monitoring by selecting non-redundant, high-impact indicators.

5. Class Imbalance Requires Special Handling

Because non-bankrupt companies dominate, standard modeling (or analytic dashboards) can be skewed. Techniques used in EDA (and later modeling) include:

Resampling (SMOTE, undersampling)

Threshold adjustment

Use of metrics like recall, precision, F1 for the minority class.

Insight: Business alert systems must balance false positives and false negatives based on risk tolerance.
